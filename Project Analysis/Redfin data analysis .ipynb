## Filter Datas For Colorado

```python
colorado = df[df['state'] == 'Colorado']
```

```python
(colorado.isna().sum())/len(colorado)*100
```

```python
# Coverting Convert Date Column Type to datetime for being able to sort values historically.
df['period_end'] = pd.to_datetime(df['period_end'])
```

```python
# Shorting value historically 
colorado = colorado.sort_values(by='period_begin')
```

```python
# Summary of the numerical variables
numerical_vars=['period_end', 'region_type', 'region',
       'city', 'state', 'property_type','homes_sold', 'sold_above_list', 'median_list_price']
colorado[numerical_vars].describe()
```

```python
#Let's Visualize the missing value
sns.heatmap(colorado.isnull(),yticklabels=False,cbar=False)
```

```python
colorado['median_sale_price'].describe()
```

```python
coloradofiltered = colorado[['period_end','region','property_type','median_sale_price','median_list_price','homes_sold','new_listings','sold_above_list', 'price_drops', 'off_market_in_two_weeks', 'parent_metro_region']]
```

```python
coloradofiltered.head()
```

```python
coloradofiltered.columns
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(coloradofiltered.corr(),annot=True,cmap = "RdYlGn")
plt.show()
```

Off market in two weeks and sold above price shows positive correlation. The more the houses are off market in two weeks the more the selling price of the houses.
During the same time period, it can be seen that all new listings are sold above the lisitng price and no significant price drop were observed as shown in the heat map. 
Also, we can see that the median list price of the houses does not make any difference to the number of houses sold showing no significant correlation between the median list price and the number the house sold.


```python
#plot the distribution of variables with most correlation
fig, axes = plt.subplots(figsize=(20,10))
axes.ticklabel_format(style='plain', axis='y')
year = pd.DatetimeIndex(coloradofiltered['period_end']).year
sns.lineplot(x=year, y=coloradofiltered['median_sale_price'], ax=axes)

```

The line graph of our dataset for Colorado shows the median sale price from the year 2012 to 2022 has been increasing.
